```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A=np.array([[1,0],[0,1]])
print(2*A)
plt.imshow(A,cmap='viridis')
```

```python
A=np.array([[1,0],[0,2]])
plt.imshow(A,cmap='viridis',vmin=0,vmax=1)
```

```python
A=np.array([[.1,.2],[.3,.4]])
fig,axes = plt.subplots(1,2, figsize=(6,3))
axes[0].imshow(A,cmap='viridis', vmin=0, vmax=1 )
axes[1].imshow(A,cmap='viridis',vmin=0, vmax=1)
plt.show()
```

```python
A=np.array([[.1,.2,.3]])
print(2*A)
plt.imshow(A,cmap='viridis')
```

```python
a_2d=np.array([[.1,.2,.3],[.4,.5,.6]])
#print(a_2d)
print(a_2d[:,1])
```

```python
fig, axes =plt.subplots(1,3,figsize=(9,3))
axes[0].imshow(a_2d,cmap='viridis',vmin=0,vmax=1)
axes[1].imshow(a_2d.T,cmap='viridis',vmin=0,vmax=1)
axes[2].imshow(a_2d+0.3,cmap='viridis',vmin=0,vmax=1)
plt.show()
print(a_2d+.1)
```

```python
fig, axes =plt.subplots(2,3,figsize=(9,6))
axes[0,0].imshow(a_2d,cmap='viridis',vmin=0,vmax=1)
axes[0,1].imshow(a_2d.T,cmap='viridis',vmin=0,vmax=1)
axes[0,2].imshow(a_2d+0.3,cmap='viridis',vmin=0,vmax=1)
axes[1,0].imshow(a_2d, cmap='gray', vmin=0, vmax=1)
axes[1,1].imshow(a_2d, cmap='Reds', vmin=0, vmax=1)
axes[1,2].imshow(a_2d, cmap='Blues', vmin=0, vmax=1)
fig.suptitle("Various Ways of Graphing 2D Tensors")
axes[0,0].set_title("Original Tensor")
axes[0,1].set_title("Transposed Tensor")
axes[0,2].set_title("Tensor + 0.3")
axes[1,0].set_title("Grayscale")
axes[1,1].set_title("Red Channel")
axes[1,2].set_title("Blue Channel")
plt.show()
```

```python
print(a_2d)
a_2d_t=a_2d>0.2
print(a_2d_t)
plt.imshow(a_2d_t,cmap='Grays')
plt.show()
```
