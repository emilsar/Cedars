# 1D Tensors

```python
import numpy as np

#1D (vector)
v=np.array([2,3,5])
print(v)
print("Shape:",v.shape)
print("Dimension:",v.ndim)
print(v*2)
print(v+2)
print("Sum:",np.sum(v))
print("Mean:",np.mean(v))
```

# 2D Tensors: Matrices


```python
#2D Arrays
M=np.array([ [2,3,5] ,[7,9,11]  ])
A=np.array([[1,2],[3,5]])
B=np.array([[2,3],[5,6]])
#print(M)
#print("Shape:",M.shape)
#print("Dimension:",M.ndim)

#Access
#The row does not matter, pick column 2
M[:,2]
#The column does not matte, pick row 1
M[0,:]
#sum up the first column
#Each axis corresponds to a dimension of movement through the array.
#Axis 0 → moves down the rows (changes the row index)
#Axis 1 → moves across the columns (changes the column index)
print(np.sum(M[:,0]))
#Sum by column:
print(np.sum(M,axis=0))
print(M.T) #Matrix Transpose
print(np.matmul(A,B))
```
